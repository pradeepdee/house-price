 treating  the null values

```python
## In the categrocial datatype we dont have the outliers 
>> now we need to know how this column is affecting the target values
>>target value=sale_price
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.countplot(df.Fireplace_Quality)
```

```python
#> we need to make data as simple as possible to prdict so minimize the level of categorical datas in the given datasets
so for this we will run annova and turkeyhsd
```

```python
## know we need to run the test to determine the correlation of Fireplace_Quality vs sales_price
##testing (wheather the Fireplace_Quality is good predicater or not)
#>>>ANOVA TEST :to check is their is a difference in categorical data 
#>> libraries
import statsmodels.api as sm
from statsmodels.formula.api import ols
#>libraries for tukey
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
##ANNOVA for Fireplace_Quality
anova1=ols('Sale_Price~Fireplace_Quality',data=df).fit()
anova1way=sm.stats.anova_lm(anova1,type=2)
print(anova1way)
 
## p(value)<0.05 it means that their is atleast 1 sub categroical data in the column casuing the some differnce against the targetvalue
```

```python
## so we will perform the tukeyhsd to determine which sub categorical data casuing the differnce
Tukey=pairwise_tukeyhsd(df.Sale_Price,df.Fireplace_Quality,alpha=0.05)
print(Tukey)
#false means it rejecting the null hypothesis
#true value casuing the differnce afainst the targetvalue
```

```python
df['Fireplace_Quality'].replace('Fa' ,'Gd',inplace=True)
```

```python
df['Fireplace_Quality'].replace('Gd' ,'Po',inplace=True)
```

```python
Tukey=pairwise_tukeyhsd(df.Sale_Price,df.Fireplace_Quality,alpha=0.05)
print(Tukey)
```

```python
df.Fireplace_Quality.value_counts()
```

```python
## after equating the sub categorical data we will finalize that the given column=Fireplace_Quality is affecting the target value so the given column is good predictor
#> Fireplace_Quality is good predictor
```

```python
## ploting the boxplot to finding the relation bwteen fireplae_quality vs gtarget_value
sns.boxplot(x='Fireplace_Quality',y='Sale_Price',data=df)
plt.title('Box plot of Fireplace_Quality vs sale_price')
```
